```python
import pandas as pd
```

```python
cols = ['basho_id', 'rikishi_id', 'elo_score', 'other_scores', 'results']

elo_scores = pd.read_csv(
    'data/processed/csv/elo_scores.csv',
    names=cols,
    )
```

```python
elo_scores
```

```python
elo_scores.sort_values(by='elo_score', ascending=False).head(25)
```

```python
g = elo_scores.groupby(by='rikishi_id') \
    .agg(elo_score=pd.NamedAgg(column='elo_score', aggfunc='max'),
         n_bouts=pd.NamedAgg(column='elo_score', aggfunc=len),
         ) \
    .sort_values(by='elo_score', ascending=False)
```

```python
g.head(15)
```

```python
g.tail(10)
```

```python
g['ratio'] = g['elo_score'] / g['n_bouts']
```

```python
g.sort_values(by='ratio').head(10)
```
